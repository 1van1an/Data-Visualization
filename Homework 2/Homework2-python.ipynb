### Part 3

The “lung_cancer_prediction_dataset.csv” provides valuable insights into lung cancer cases, risk
factors, smoking trends, and healthcare access across 25 of the world's most populated
countries. It includes 220,632 individuals with details on their age, gender, smoking history,
cancer diagnosis, environmental exposure, and survival rates. The dataset is useful for medical
research, predictive modeling, and policymaking to understand lung cancer patterns globally.

The “global_air_pollution_dataset.csv” provides information about the contamination of the
indoor or outdoor environment by any chemical, physical or biological agent that modifies the
natural characteristics of the atmosphere. It includes 23,463 rows with data presented in the
following columns: Country, City, AQI Value, AQI Category, CO AQI Category, Ozone AQI Value,
Ozone AQI Category, NO2 AQI Value, NO2 AQI Category, PM2.5 AQI Value, PM2.5 AQI Category.

1. Create a Boxplot of Lung Cancer Deaths Distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

data_lung_cancer = pd.read_csv('lung_cancer_prediction_dataset.csv')
data_air_pollution = pd.read_csv('global_air_pollution_dataset.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data_lung_cancer['Annual_Lung_Cancer_Deaths'])
plt.title('Boxplot of Lung Cancer Deaths Distribution')
plt.ylabel('Number of Deaths')
plt.show()
```

2. Create a Histogram of PM2.5 AQI Values.

```python
plt.figure(figsize=(12, 8))
plt.hist(data_air_pollution['AQI_Value'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of AQI Values Across Cities')
plt.xlabel('AQI Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

3. Create a Density Plot of the Lung Cancer Mortality Rate. 

```python
density = data_lung_cancer['Mortality_Rate'].plot(kind='density', color='blue', linewidth=2)
x, y = density.get_lines()[0].get_data()
plt.fill_between(x, y, color="blue", alpha=0.3)

plt.title('Density Plot of Lung Cancer Mortality Rate')
plt.xlabel('Mortality Rate')
plt.grid(True)
plt.show()
```

1. Use the matplotlib library for this graph.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('global_air_pollution_dataset.csv')

pm25_values = df["PM2.5_AQI_Value"].dropna()

plt.figure(figsize=(10, 6))
sns.histplot(pm25_values, bins=30, kde=True, color="blue", stat="density", alpha=0.5)

sns.kdeplot(pm25_values, color="darkred", linewidth=2)

plt.xlabel("PM2.5 AQI Value")
plt.ylabel("Density")
plt.title("PM2.5 AQI Distribution with Density Overlay")

plt.legend(["Density Plot (KDE)", "Histogram"], loc="upper right")

plt.show()
```
